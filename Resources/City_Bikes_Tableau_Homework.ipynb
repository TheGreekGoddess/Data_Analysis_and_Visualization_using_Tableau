# City Bike Data trends

```python
# Import the various dependencies and setup
import pandas as pd
```

```python
# Map the path where the input csv file is located
YTD = "Resources_YTD_2020\Citi_Bike_Data_Analytics_YTD_2020.csv"
```

```python
# Specify the delimiter to read and store the csv file into a Pandas DataFrame
YTD = pd.read_csv(YTD)
```

```python
# Display the YTD DataFrame
YTD.head(2)
```

```python
# Map the path where the input csv file is located
YOY = "Resources_YOY_July\Citi_Bike_Data_Analytics_YOY_July.csv"
```

```python
# Specify the delimiter to read and store the csv file into a Pandas DataFrame
YOY = pd.read_csv(YOY)
```

```python
# Display the YOY DataFrame
YOY.head(2)
```

```python
# Combine the two DataFrames into a single DataFrame
Citi_Bike_Data = YTD.append([YOY], ignore_index = True)
Citi_Bike_Data.head(2)
```

```python
# Save the combined DataFrame to CSV
Citi_Bike_Data.to_csv("Citi_Bike_Data.csv", index = False, header = True)
```
